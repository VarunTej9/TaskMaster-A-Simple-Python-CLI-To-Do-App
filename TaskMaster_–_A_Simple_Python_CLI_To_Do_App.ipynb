```python
import json
import os

FILE_NAME = "todo_list.json"

# Load tasks from file
if os.path.exists(FILE_NAME):
    with open(FILE_NAME, 'r') as file:
        todo_list = json.load(file)
else:
    todo_list = []

def save_tasks():
    with open(FILE_NAME, 'w') as file:
        json.dump(todo_list, file)

def show_menu():
    print("\nTo-Do List Options:")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Remove Task")
    print("4. Mark Task as Completed")
    print("5. Edit Task")
    print("6. Exit")

def view_tasks():
    if not todo_list:
        print("No tasks yet!")
    else:
        print("\nYour Tasks:")
        for idx, task in enumerate(todo_list, start=1):
            status = "[Yes]" if task['completed'] else "[ ]"
            print(f"{idx}. {status} {task['task']}")

def add_task():
    task_text = input("Enter the task: ")
    task = {"task": task_text, "completed": False}
    todo_list.append(task)
    save_tasks()
    print("Task added!")

def remove_task():
    view_tasks()
    try:
        index = int(input("Enter task number to remove: ")) - 1
        removed = todo_list.pop(index)
        save_tasks()
        print(f"Removed: {removed['task']}")
    except (ValueError, IndexError):
        print("Invalid task number.")

def mark_completed():
    view_tasks()
    try:
        index = int(input("Enter task number to mark completed: ")) - 1
        todo_list[index]['completed'] = True
        save_tasks()
        print(f"Marked as completed: {todo_list[index]['task']}")
    except (ValueError, IndexError):
        print("Invalid task number.")

def edit_task():
    view_tasks()
    try:
        index = int(input("Enter task number to edit: ")) - 1
        new_text = input("Enter new task description: ")
        todo_list[index]['task'] = new_text
        save_tasks()
        print("Task updated!")
    except (ValueError, IndexError):
        print("Invalid task number.")

while True:
    show_menu()
    choice = input("Choose an option (1–6): ")

    if choice == "1":
        view_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        remove_task()
    elif choice == "4":
        mark_completed()
    elif choice == "5":
        edit_task()
    elif choice == "6":
        print("Goodbye!")
        break
    else:
        print("Invalid option. Please choose 1–6.")
```
